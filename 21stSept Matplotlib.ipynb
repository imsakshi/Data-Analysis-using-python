```python
a_array=np.linspace(0,10,20)
```

```python
print(a_array)
```

```python
plt.title("Pressure Sensor")
plt.plot(a_array,color="g",marker = "o",linestyle = "--",label="Pres Valt1")
plt.plot(np.linspace(10,0,20),color="r",marker="o",label="Pres Valt2")
plt.xlabel("Time")
plt.ylabel("Pressure")
plt.grid()
plt.legend()
```

```python
import matplotlib.pyplot as plt
```

```python
sales_dict={'apple':100,'banana':95,'cherry':85,'Limes':70,'Orange':60}
```

```python
bar_list=plt.barh(list(sales_dict.keys()),list(sales_dict.values()),color="Orange")
bar_list[0].value = 75 
bar_list[0].set_color("75")
bar_list[-1].set_color("k")
plt.xlabel("Fruits")
plt.grid()
plt.axvline(10)
plt.show()

```
