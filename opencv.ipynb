```python
h1 = 300
w1 = 300
dimension = (w1,h1)
resized_image = cv2.resize(img,dimension,interpolation = cv2.INTER_AREA)
cv2.imshow("resized",resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
(h1,w1) = img.shape
center = (w1/2,h1/2)
mat = cv2.getRotationMatrix2D(center,180,1.0)
rotated_image = cv2.warpAffine(img,mat,(w1,h1))
cv2.imshow("rotated",rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
cropped_image = img[100:170,250:350]
cv2.imshow("cropped",cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
#Add different shapes
img = cv2.line(img,(0,0),(255,255),(63,127,0),10)
img = cv2.arrowedLine(img,(0,255),(255,255),(255,0,0),10)
img = cv2.rectangle(img,(384,0),(510,128),(0,0,255),10)
img = cv2.circle(img,(447,63),63,(0,255,0))
img = cv2.circle(img,(250,300),100,(0,0,255))
font = cv2.FONT_HERSHEY_COMPLEX
img = cv2.putText(img,"Hi, i am elle ",(10,500),font,4,(0,0,127))

cv2.imshow("image",img)#black-white image
cv2.waitKey(0)
cv2.destroyAllWindows()
```
